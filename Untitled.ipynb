```python
all_files=[]
for yr in range(2000,2014+1):
    all_files.append(pd.read_csv('/Users/Parshu/Desktop/tableau/names/yob{}.txt'.format(yr),names=['name','sex','Numbers']))
                   
# add a column for a year and fill it
    all_files[-1]['Year']=yr
all_years=pd.concat(all_files)
all_years.head(4)
all_years.tail(4)
```

```python
#index the data on gender,name and year

allYrInx=all_years.set_index(['sex','name','Year']).sort_index()
allYrInx
```

```python
def plotgraph(sex,name):
    data = allYrInx.loc[sex,name]

    mlp.plot(data.index,data.value)
```
